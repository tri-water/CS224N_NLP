```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from sklearn.decomposition import PCA
from gensim.test.utils import datapath, get_tmpfile
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
```

```python
# Get full path for file fname in test data directory placed in this module directory.
# Usually used to place corpus to test_data directory
glove_file = datapath(r'C:/Users/gao_x/Desktop/Stanford CS224n/Lecture 1/glove.6B/glove.6B.100d.txt')
# Get full path to file suffix in temporary folder. 
# This function doesn't create file (only generate unique name).
# Also, it may return different paths in consecutive calling.
word2vec_glove_file = get_tmpfile("glove.6B.100d.word2vec.txt")
# Convert the first file in GloVe format to word2vec format and write it to the second file
glove2word2vec(glove_file, word2vec_glove_file)
```

```python
# Create a model for similarity look-ups
model = KeyedVectors.load_word2vec_format(word2vec_glove_file)
```

```python
model.most_similar('obama')
```

```python
model.most_similar('banana')
```

```python
model.most_similar(negative='banana')
```

```python
result = model.most_similar(positive=['woman', 'king'], negative=['man'])
print("{}: {:.4f}".format(*result[0]))
```

```python
def analogy(x1, x2, y1):
    result = model.most_similar(positive=[y1, x2], negative=[x1])
    return result[0][0]
```

```python
analogy('japan', 'japanese', 'australia')
```

```python
analogy('australia', 'beer', 'france')
```

```python
analogy('obama', 'clinton', 'reagan')
```

```python
analogy('tall', 'tallest', 'long')
```

```python
analogy('good', 'fantastic', 'bad')
```

```python
def display_pca_scatterplot(model, words=None, sample=0):
    if words == None:
        if sample > 0:
            words = np.random.choice(list(model.vocab.keys()), sample)
        else:
            words = [word for word in model.vocab]
        
    word_vectors = np.array([model[w] for w in words])
    
    twodim = PCA().fit_transform(word_vectors)[:, :2]
    
    plt.figure(figsize=(6, 6))
    plt.scatter(twodim[:, 0], twodim[:, 1], edgecolors='k', c='r')
    for word, (x, y) in zip(words, twodim):
        plt.text(x + 0.05, y + 0.05, word)
```

```python
display_pca_scatterplot(
    model,
    ['coffee', 'tea', 'beer', 'wine', 'brandy', 'rum', 'champagne', 'water',
     'spaghetti', 'borscht', 'hamburger', 'pizza', 'falafel', 'sushi', 
     'meatballs', 'dog', 'horse', 'cat', 'monkey', 'parrot', 'koala', 
     'lizard','frog', 'toad', 'monkey', 'ape', 'kangaroo', 'wombat', 'wolf', 
     'france', 'germany', 'hungary', 'luxembourg', 'australia', 'fiji', 
     'china', 'homework', 'assignment', 'problem', 'exam', 'test', 'class',
     'school', 'college', 'university', 'institute'])
```

```python
display_pca_scatterplot(model, sample=300)
```
